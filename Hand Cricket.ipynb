```python
import random

score = 0
compscore = 0

def toss():
    global score, compscore
    
    Toss = input('Enter either odd or even: ')
    playerinp = int(input('Enter a number from 1 to 6: '))
    print(f'Player chose {playerinp}')
    computerinp = random.randint(1, 6)
    print(f'Computer chose {computerinp}')

    if Toss.lower() == 'even':
        if ((playerinp + computerinp) % 2) == 0:
            print('Player won the toss')
            playerchoice = input('Choose either bat or bowl: ')
            if playerchoice.lower() == 'bat':
                print("Player chose batting")
                firstbatting()
                secondbatting()
            elif playerchoice.lower() == 'bowl':
                print('Player chose bowling')
                secondbatting()
                firstbatting()
            else:
                print('Invalid input')
                toss()
        else:
            print('Computer won the toss')
            computerchoice = random.choice(['bat', 'bowl'])
            if computerchoice == 'bat':
                print('Computer chose batting')
                secondbatting()
                firstbatting()
            elif computerchoice == 'bowl':
                print('Computer chose bowling')
                firstbatting()
                secondbatting()
    elif Toss.lower() == 'odd':
        if ((playerinp + computerinp) % 2) != 0:
            print('Player won the toss')
            playerchoice = input('Choose either bat or bowl: ')
            if playerchoice.lower() == 'bat':
                print("Player chose batting")
                firstbatting()
                secondbatting()
            elif playerchoice.lower() == 'bowl':
                print('Player chose bowling')
                secondbatting()
                firstbatting()
            else:
                print('Invalid input')
                toss()
        else:
            print('Computer won the toss')
            computerchoice = random.choice(['bat', 'bowl'])
            if computerchoice == 'bat':
                print('Computer chose batting')
                secondbatting()
                firstbatting()
            elif computerchoice == 'bowl':
                print('Computer chose bowling')
                firstbatting()
                secondbatting()
    else:
        print('Invalid input')
        toss()

def firstbatting():
    global score
    score = 0
    while True:
        print('Player is ready to bat')
        playerturn = int(input('Enter a number from 1 to 6: '))
        print(f'Player chose {playerturn}')
        print('Computer is ready to bowl')
        computerturn = random.randint(1, 6)
        print(f'Computer chose {computerturn}')
        score += playerturn
        print(f'The score is {score}')
        if computerturn == playerturn:
            print('Player loses, computer dismissed player')
            break

def secondbatting():
    global compscore
    compscore = 0
    while True:
        print('Computer is ready to bat')
        computerturn = random.randint(1, 6)
        print(f'Computer chose {computerturn}')
        print('Player is ready to bowl')
        playerturn = int(input('Enter a number from 1 to 6: '))
        print(f'Player chose {playerturn}')
        compscore += computerturn
        print(f'The score is {compscore}')
        if computerturn == playerturn:
            print('Computer loses, player dismissed computer')
            break
toss()
```
